```python
# Dependencies
import numpy as np
import datetime as dt
from datetime import date
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# Import the sqlalchemy extract function.
from sqlalchemy import extract

# 1. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 

# Filter
month = "06"

# Variable to store the results of the query
results = []

# Perform the query
total_rows = session.query(func.count(Measurement.date)).all()
print (f'there are {total_rows} in the database')
results = session.query(Measurement.date, Measurement.tobs).filter(func.strftime("%m", Measurement.date) == month)
```

```python
# 2. Convert the June temperatures to a list.
results.all()
```

```python
# 3. Create a DataFrame from the list of temperatures for the month of June. 
jun_df = pd.DataFrame(results, columns=['date', 'temperature'])
jun_df.set_index(jun_df['date'], inplace=True)
jun_df = jun_df.sort_index()
jun_df = jun_df.drop('date', axis=1)
jun_df
```

```python
# 4. Calculate and print out the summary statistics for the June temperature DataFrame.
jun_df.describe()
```

```python
# Additional QC, create a bar plot for the June tempratures
jun_df.plot.hist(bins=15)
plt.tight_layout()
plt.title('June Temperatures (°F)')
plt.xlim([60,90])
plt.ylim([0,400])
```
